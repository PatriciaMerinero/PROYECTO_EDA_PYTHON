# 1.3. ESTADÍSTICAS DESCRIPTIVAS

## Normalizo el target a yes/no y creo binario

```python
df = df_full.copy()
y_norm = df['y'].astype(str).str.strip().str.lower().replace({'sí':'yes','si':'yes'})
df['y_bin'] = (y_norm == 'yes').astype(int)

print("Shape:", df.shape)
print("Tipos de columnas:\n", df.dtypes.value_counts(), "\n")
```

## Resumen de nulos por columna (conteo y %)

```python
null_tbl = pd.DataFrame({
    'nulos': df.isna().sum(),
    '%_nulos': (df.isna().mean()*100).round(2),
    'n_unique': df.nunique(dropna=True)
}).sort_values('%_nulos', ascending=False)
display(null_tbl.head(20))
```

## Numéricas: describe ampliado + info de nulos y únicos.

```python
num_cols = df.select_dtypes(include='number').columns.drop(['y_bin'], errors='ignore')
desc_num = df[num_cols].describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]).T
desc_num['n_unique']   = df[num_cols].nunique()
desc_num['nulos']      = df[num_cols].isna().sum()
desc_num['%_nulos']    = (df[num_cols].isna().mean()*100).round(2)
display(desc_num.head(12))
```

## Categóricas: cardinalidad, moda y % nulos

```python
cat_cols = df.select_dtypes(exclude='number').columns.tolist()
# quita IDs del resumen si los tienes
cat_cols = [c for c in cat_cols if c.lower() not in ('id','id_')]
desc_cat = pd.DataFrame({
    'n_unique': df[cat_cols].nunique(),
    'top':      [df[c].mode(dropna=True).iloc[0] if df[c].dropna().size else None for c in cat_cols],
    'top_freq': [df[c].value_counts(dropna=True).iloc[0] if df[c].dropna().size else 0 for c in cat_cols],
    '%_nulos':  (df[cat_cols].isna().mean()*100).round(2)
}).sort_values('n_unique', ascending=False)
display(desc_cat.head(20))
```

## Numéricas por target (yes/no): medias, medianas, desviación

```python
grp = df.groupby((y_norm == 'yes').map({True:'yes', False:'no'}))
num_by_y = grp[num_cols].agg(['count','mean','median','std']).round(20)
display(num_by_y.head(10))
```

## Tasas de conversión por categóricas (filtra categorías raras con min_count)

```python
def rate_by(col, min_count=50):
    g = df.groupby(col)
    out = pd.DataFrame({
        'n': g.size(),
        'yes_%': (g['y_bin'].mean()*100).round(2)
    }).query('n >= @min_count').sort_values('yes_%', ascending=False)
    return out

for col in ['job','education','marital','housing','loan','contact','poutcome','region','country']:
    if col in df.columns:
        print(f"\n% YES por {col} (min 50 obs)")
        display(rate_by(col, min_count=50).head(10))
```

## 1.4. RESUMEN ESTADÍSTICO

```python
def resumen_estadistico(df, target='y', drop_ids=('ID','id_')):
    # --- setup
    cols_ok = [c for c in df.columns if c not in drop_ids]
    dfx = df[cols_ok].copy()

    # --- tamaño y tipos
    shape = dfx.shape
    dtype_counts = dfx.dtypes.value_counts().astype(int)

    # --- nulos
    total_null_pct = round(dfx.isna().sum().sum() / (dfx.size) * 100, 2)
    top_null_cols = (dfx.isna().mean()*100).sort_values(ascending=False).head(10).round(2)

    # --- numéricas
    num_cols = dfx.select_dtypes(include='number').columns
    desc_num = dfx[num_cols].describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]).T
    desc_num['n_unique'] = dfx[num_cols].nunique()
    desc_num['%_nulos'] = (dfx[num_cols].isna().mean()*100).round(2)

    # --- categóricas
    cat_cols = dfx.select_dtypes(exclude='number').columns
    desc_cat = pd.DataFrame({
        'n_unique': dfx[cat_cols].nunique(),
        'top':      [dfx[c].mode(dropna=True).iloc[0] if dfx[c].dropna().size else None for c in cat_cols],
        'top_freq': [dfx[c].value_counts(dropna=True).iloc[0] if dfx[c].dropna().size else 0 for c in cat_cols],
        '%_nulos':  (dfx[cat_cols].isna().mean()*100).round(2)
    }).sort_values('n_unique', ascending=False)

    # --- target (si existe)
    dist_target = None
    num_by_target = None
    if target in df.columns:
        y_norm = df[target].astype(str).str.strip().str.lower().replace({'sí':'yes','si':'yes'})
        dist_target = (y_norm.value_counts(normalize=True)*100).round(2)
        num_by_target = df.groupby(y_norm)[num_cols].agg(['count','mean','median','std']).round(2)

    return {
        'shape': shape,
        'dtype_counts': dtype_counts,
        'total_null_pct': total_null_pct,
        'top_null_cols': top_null_cols,
        'desc_num': desc_num,
        'desc_cat': desc_cat,
        'dist_target': dist_target,
        'num_by_target': num_by_target
    }

# ---- ejecutar
summary = resumen_estadistico(df_full, target='y')

print("Tamaño (filas, columnas):", summary['shape'])
print("\nColumnas por tipo:\n", summary['dtype_counts'])
print("\n% nulos total:", summary['total_null_pct'])
print("\nTop 10 columnas con más nulos (%):\n", summary['top_null_cols'])

display(summary['desc_num'].head(12))  # primeras 12 numéricas
display(summary['desc_cat'].head(20))  # top 20 categóricas por cardinalidad

if summary['dist_target'] is not None:
    print("\nDistribución del target ( % ):")
    print(summary['dist_target'])
    display(summary['num_by_target'].head(10))

```
